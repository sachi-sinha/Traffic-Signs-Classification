# Load Data

```python
import cv2
import numpy as np
import os
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
EPOCHS = 10
IMG_SIZE = (30, 30)
NUM_CATEGORIES = 43
TEST_SIZE = 0.4
```

```python
path = './gtsrb'
```

<p style = "font-size: 17px">To ensure that all images have the same size, we resize the images to 30 x 30. </p>

```python
# Load Data
images, labels = list(), list()
for folder in os.listdir(path):
        folder_path = os.path.join(path, folder)
        if os.path.isdir(folder_path):
            for image_file in os.listdir(folder_path):
                image = cv2.imread(os.path.join(folder_path, image_file), 1)
                image = cv2.resize(image, IMG_SIZE, interpolation=cv2.INTER_AREA)
                images.append(image)
                labels.append(int(folder))
```

# Data Summary

```python
labels = tf.keras.utils.to_categorical(labels)
x_train, x_test, y_train, y_test = train_test_split(
        np.array(images), np.array(labels), test_size=TEST_SIZE
    )
```

```python
print("Number of training examples =", len(x_train))
print("Number of testing examples =", len(x_test))
print("Image data shape =", images[0].shape)
print("Number of labels =", len(os.listdir(path)))
```

## Visualize the images in German Traffic Sign Recognition Benchmark Dataset

```python
display_images = [0 for i in range(len(os.listdir(path)))]
for folder in os.listdir(path):
        folder_path = os.path.join(path, folder)
        if os.path.isdir(folder_path):
            image = cv2.imread(os.path.join(folder_path, os.listdir(folder_path)[5]), 1)
            display_images[int(folder)] = image
for i in range(len(display_images)):
    plt.imshow(display_images[i])
    plt.title("Class: " + str(i))
    plt.show()
```

# Developing the Model

<ul style = "font-size: 17px">
    <li>The project applies an input convolutional layer of 32 filters and a kernal of 3 X 3 matrix. The input layer is a 30 X 30 matrix. </li>
    <li>It then applies max pooling with a pool size of 2 X 2 matrix to the model, which picks out the maximum value in the pool region. </li>
    <li>The first combination of convolution and pooling helps in identifying low level features like edges, curves and shapes in the data. </li>
    <li>The project again applies the combination of convolution and pooling to identify objects in the data. </li>
    <li>It applies two hidden layers - one with 128 units and 0.2 dropout and second with 256 units and 0.5 dropout. </li>
    <li>Dropout values have been used after several hit and trials to make sure that the model is neither overfitting nor underfitting. </li>
    <li>It finally applies an output layer with 43 units, corresponding to the 43 different traffic signs in the data</li>
</ul>

```python
# Create a Convolutional Neural Network
model = tf.keras.models.Sequential([
    
    # Convolutional layer
    tf.keras.layers.Conv2D(
        32, (3, 3), activation = "relu", input_shape = (30, 30, 3)
    ),
    
    # Max-pooling layer
    tf.keras.layers.MaxPooling2D(pool_size = (2, 2)),
    
    # Convolutional layer
    tf.keras.layers.Conv2D(
        64, (3, 3), activation = "relu"
    ),
    
    # Max-pooling layer
    tf.keras.layers.MaxPooling2D(pool_size = (2, 2)),
    
    # Flatten units
    tf.keras.layers.Flatten(),
    
    # Add 2 hidden layers with dropout
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    
    # Add an output layer
    tf.keras.layers.Dense(43, activation = "softmax")
])
```

## Train the Neural Network

```python
# Train a Neural Network
model.compile(
    optimizer = "adam",
    loss = "categorical_crossentropy",
    metrics = ["accuracy"]
)
```

## Fit the Model

```python
# Fit Model
model.fit(x_train, y_train, epochs=EPOCHS)
```

## Evaluate the Neural Network

```python
# Evaluate neural network performance
model.evaluate(x_test,  y_test, verbose=2)
```

# Saving the Model

```python
model.save("model.h5")
```
